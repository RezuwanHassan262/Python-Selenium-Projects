```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless') #we want it to be headless
options.add_argument('--no-sandbox') #other options that are critical for it to actually work properly
options.add_argument('--disable-dev-shm-usage')
```

```python
site = 'http://ahdb.org.uk/cereals-oilseeds-markets'
wd = webdriver.Chrome('chromedriver',options = options) #Using chromedriver and the options we stated in the above cell
wd.get(site) #generates updated html
html = wd.page_source #grabbing the html from the page source
df = pd.read_html(html) #loading the grabbed html into a dataframe
```
